## Plot heampmap

```python
# imports
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from utils import get_matrix_change
```

```python
# make the calculus
letter = 'BTC-USD'
letter = 'SOL1-USD'
letter = 'ADA-USD'

df = get_matrix_change(letter)
```

```python
# plot
plt.figure()
plt.title(f"% Changes of {letter} probability")

plot_conf = {
    "vmin":0, 
    "vmax":100, 
    "annot":True, 
    "linewidths":.5, 
    "cmap":"YlGnBu"
}

sns.heatmap(df, **plot_conf)
```
